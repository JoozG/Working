```python
#----------------------------------------FUNKCJE-----------------------------------------
#Funckja# Utworzenie tabeli ze stringa
def totable(string1):
    table=[]
    
    for char in string1:
        table.append(char)

    return table

#Funckja# Utworzenie tabeli kodonów ze stringa 
def tocodone(string1):
    
    table = [string1[i:i+3] for i in range(0, len(string1), 3)]
    
    return table

#Funckja# Zliczanie wystąpień poszczególnych nukleotydów
def nuc_count (table):
    count_A=0
    count_C=0
    count_G=0
    count_T=0
    count_undefined=0
    for a in table:
        temp=a

        match temp:
            case "A":
                count_A+=1
            case "C":
                count_C+=1
            case "G":
                count_G+=1
            case "T":
                count_T+=1
            case _:
                count_undefined+=1
    print("Ilość wystąpień adenozyny: ",count_A)
    print("Ilość wystąpień cytozyny: ",count_C)
    print("Ilość wystąpień guaniny: ",count_G)
    print("Ilość wystąpień tyrozyny: ",count_T)
    print("Ilość elementów niezdefiniowanych: ",count_undefined)

#Funckja# Transkrypcja z DNA na korespondujące RNA
def trans_DNA_to_RNA(table):
    trans=[]
    for a in table:
        if a == "T":
            trans.append("U")
        elif a in ("A", "G", "C"):
            trans.append(a)

    return trans

#Funckja# Znalezienie komplementarnej nici DNA
def second_strand_DNA(table):
    ss_DNA=[]

    for a in table:
        match a:
            case "A":
                ss_DNA.append("T")
            case "C":
                ss_DNA.append("G")
            case "G":
                ss_DNA.append("C")
            case "T":
                ss_DNA.append("A")
    ss_DNA.reverse()
    return ss_DNA
#---------------------------------------

#Funckja# Wczytywanie pliku z sekwencjami nukleotydów w formacie FASTA
def fasta_load(filename):
    auth_ext=[".fasta", ".fas", ".fa", ".fna", ".ffn", ".faa", ".mpfa", ".frn"]
    names=[]
    nuc_seqs=[]
    current_sequence=""

    try:

        with open(filename, 'r') as file:
            lines = file.readlines()

    except FileNotFoundError:
        print("Plik nie został odnaleziony")

    for line in lines:

        line = line.strip()

        if line.startswith(">"):
            names.append(line)
            if current_sequence:  # Sprawdzenie, czy istnieje już sekwencja do zapisania
                nuc_seqs.append(current_sequence)
                current_sequence = ""  # Wyczyszczenie sekwencji

        else:
            current_sequence += line
    
    # Dodanie ostatniej sekwencji, jeśli została znaleziona na końcu pliku
    if current_sequence:
        nuc_seqs.append(current_sequence)

    output = tuple(zip(names,nuc_seqs))

    return output
#----------------------------------

#Funckja# Wyliczanie zawartość GC
def GC_cont(tuple1):
    GC=[]
    names=[]

    for pair in tuple1:
        length=len(pair[1])
        count_C=pair[1].count("C")
        count_G=pair[1].count("G")
        percentage=round(100*(count_C+count_G)/length,6)
        GC.append(percentage)
        names.append(pair[0])

    output = tuple(zip(names,GC))

    return output

#Funkcja# Zwraca maksimum krotki, ustryktuyzowanej następująca krokta[0]=nazwa, krotka[1]=wartość
def max_of_tuple(tuple):
    max_record = max(tuple, key=lambda x: x[1])
    return max_record

#Funkcja#
def ham_dist(string1, string2):
    dist=0
    n=min(len(string1), len(string2))-1

    for i in range(0,n):
        if string1[i]!=string2[i]:
            dist+=1
        else:
            continue

    return dist

#Funkcja# Konwersja ze stringa do postaci 
def RNA_to_protein(string1):

    table1=tocodone(string1) #Tabela zawierająca kodony jako elementy

    codone_dict = {
    "Ala": ["GCU", "GCC", "GCA", "GCG"],
    "Ile": ["AUU", "AUC", "AUA"],
    "Arg": ["CGU", "CGC", "CGA", "CGG", "AGA", "AGG"],
    "Leu": ["CUU", "CUC", "CUA", "CUG", "UUA", "UUG"],
    "Asn": ["AAU", "AAC", "GAU", "GAC"],
    "Lys": ["AAA", "AAG"],
    "Asp": ["GAU", "GAC"],
    "Met": ["AUG"],
    "Phe": ["UUU", "UUC"],
    "Cys": ["UGU", "UGC"],
    "Pro": ["CCU", "CCC", "CCA", "CCG"],
    "Gln": ["CAA", "CAG", "GAA", "GAG"],
    "Ser": ["UCU", "UCC", "UCA", "UCG", "AGU", "AGC"],
    "Glu": ["GAA", "GAG"],
    "Thr": ["ACU", "ACC", "ACA", "ACG"],
    "Trp": ["UGG"],
    "Gly": ["GGU", "GGC", "GGA", "GGG"],
    "Tyr": ["UAU", "UAC"],
    "His": ["CAU", "CAC"],
    "Val": ["GUU", "GUC", "GUA", "GUG"],
    #"START": ["AUG", "CUG", "UUG"], #Z jakiegoś względu nie chce działać przy zadeklarowanym kodonie start
    "STOP": ["UAA", "UGA", "UAG"]
}
    amacids_dict = {
    "Ala": "A", "Ile": "I", "Arg": "R", "Leu": "L", "Asn": "N", 
    "Lys": "K", "Asp": "D", "Met": "M", "B": "B", "Phe": "F", 
    "Cys": "C", "Pro": "P", "Gln": "Q", "Ser": "S", "Glu": "E", 
    "Thr": "T", "Z": "Z", "Trp": "W", "Gly": "G", "Tyr": "Y", 
    "His": "H", "Val": "V", "START": "|START|", "STOP": "|STOP|"
    }
    print (table1)

    table2=[] #Tabela zawierająca skrótową nazwę aminokwasów jako elementy
    table3=[] #Tabela zawierająca aminokwasy oznaczone literami, jako elementy

    for codone in table1:
        for key, value_dict in codone_dict.items():
            if codone in value_dict:
                table2.append(key)
    
    return(print(table2))

    for aminoacid in table2:
        table3.append(amacids_dict[aminoacid])
    
    #print (table3)


#Funkcja# Znajdywanie zadeklarowanego podciągu w stringu
def find_substring(string1, substring):
    occur=[]
    pointer=string1.find(substring)

    while pointer !=-1:
        occur.append(pointer)
        pointer = string1.find(substring, pointer+1)
    if occur:
        return occur
    else:
        return 0

def consensus_DNA(tuple1):
    import numpy as np

    count_list=[]

    A_counter=[]
    C_counter=[]
    G_counter=[]
    T_counter=[]

    nuc_list=[]

    for sublist in tuple1:
        nuc_list.append(sublist[1])

    nuc_array=np.array([list(strand) for strand in nuc_list])

    for i in range (nuc_array.shape[1]):
        temp=(nuc_array[:,i])  

        A_counter.append(np.count_nonzero(temp=="A"))
        C_counter.append(np.count_nonzero(temp=="C"))
        G_counter.append(np.count_nonzero(temp=="G"))
        T_counter.append(np.count_nonzero(temp=="T"))

        i+=1

    count_list.append(A_counter)
    count_list.append(C_counter)
    count_list.append(T_counter)
    count_list.append(G_counter)   

    return count_list
#-------------------------


#FUNKCJA# Najdłuższy podciąg z 2 stringów
def longest_common_substrings(string1, string2):
    r = len(string1)
    n = len(string2)
    queue_list = np.zeros((r + 1, n + 1), dtype=int)
    z = 0
    substr_list = []

    for i in range(1, r + 1):
        for j in range(1, n + 1):
            if string1[i - 1] == string2[j - 1]:
                if i == 1 or j == 1:
                    queue_list[i][j] = 1
                else:
                    queue_list[i][j] = queue_list[i - 1][j - 1] + 1
                if queue_list[i][j] > z:
                    z = queue_list[i][j]
                    substr_list = {string1[i - z:i]}
                elif queue_list[i][j] == z:
                    substr_list.add(string1[i - z:i])
            else:
                queue_list[i][j] = 0
    
    return substr_list
#----------------------

#FUNKCJA# Najdłuższy podciąg z wielu stringów
def longest_common_substring(strings_list, common_strings_list):   
    for string in strings_list:
        for common_string in common_strings_list:
            print(find_substring(string, common_string))
#--------------------
```
